```python
import tensorflow as tf 
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#load the model and test
```

```python
loaded_model = load_model('ann_model_new.h5')
```

```python
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder = pickle.load(f)
with open('hot_encoder_geography.pkl', 'rb') as f:
    onehot_encoder = pickle.load(f)
```

```python
input_data = {
    'CreditScore': 619.00,
    'Gender': 'Female',
    'Age': 42.00,
    'Tenure': 2.00,
    'Balance': 0.00,
    'NumOfProducts': 1.00,
    'HasCrCard': 1.00,
    'IsActiveMember': 1.00,
    'EstimatedSalary': 101348.88,
    'Geography'  :        'France'	
  
}
```

```python
## encode one caterogical variables
from sklearn.preprocessing import LabelEncoder
input_data['Gender'] = label_encoder.transform([input_data['Gender']])[0]
```

```python
input_data['Gender']
```

```python
hot_encoded_data = onehot_encoder.transform([[input_data['Geography']]])
print(hot_encoded_data)
```

```python
hot_encoded_data
```

```python
get_encoded_df = pd.DataFrame(hot_encoded_data, columns = onehot_encoder.get_feature_names_out(['Geography']))
```

```python
get_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
```

```python
final_input = pd.concat([input_df.reset_index(drop=True), get_encoded_df.reset_index(drop=True)], axis=1)
```

```python
final_input
```

```python
final_input.drop(['Geography'], axis=1, inplace=True)
```

```python
input_df.head()
```

```python
input_scaled = scaler.transform(final_input)
```

```python
input_scaled
```

```python
prediction = loaded_model.predict(input_scaled)
```
